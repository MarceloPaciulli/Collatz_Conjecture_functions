```python

import numpy as np
import matplotlib.pyplot as plt
from functools import reduce
import matplotlib as mpl
import seaborn as sns
import warnings


#suppress warnings 'return 1/(1 + np.exp(-x))'
warnings.filterwarnings('ignore')


# Apply the default theme
sns.set_theme()


def f(n):
    return n/pow(np.sqrt(2),n)

def l(n):
    x=[]
    for i in n:
        x.append((3*i+1)/8**i)
    return x

def g(n):
    x = []
    for i in n:
        x.append(np.math.factorial(i/2)*reduce(int.__mul__, range(i-1, 0, -2))/np.math.gamma(i))
    return x

def h(n):
    x = []
    for i in n:
        x.append(np.math.factorial(3*i+1)*reduce(int.__mul__, range(6*i+1, 0, -2))/np.math.gamma(6*i+2))
    return x

def k(n):
    x = []
    for i in n:
        x.append(pow(2,-(i/2)-1)*(2-i*np.log(2)))
    return x
    

#print(reduce(int.__mul__, range(4, 0, -2)))


#print(k([4/np.log(2)]))
X = np.r_[0:30: 1000j].reshape((-1, 1))
X1 = [x for x in range(0,30,2)]

X2 = np.r_[0:30: 1000j].reshape((-1, 1))
X3 = [x for x in range(1,29,2)]

fig = plt.gcf()
fig.set_size_inches(15, 11)

ax = plt.subplot(2,2,1)
ax.get_xaxis().set_minor_locator(mpl.ticker.AutoMinorLocator())
ax.get_yaxis().set_minor_locator(mpl.ticker.AutoMinorLocator())
ax.grid(b=True, which='major', color='black', linewidth=0.3, alpha=0.5)
ax.grid(b=True, which='minor', color='black', linewidth=0.3, alpha=0.5)
ax.plot(X,f(X),color="blue")
ax.plot(X1[1:],g(X1[1:]), 'bo',color="red", markersize=8)
ax.plot(X1,f(X1), '*',color="green")
#another function
plt.plot(X2,l(X2),color="orange")
plt.plot(X3,h(X3), 'bo',color="red", markersize=8)
plt.plot(X3,l(X3), 'bo',color="green")
#####################################
ax = plt.subplot(2,2,2)
ax.get_xaxis().set_minor_locator(mpl.ticker.AutoMinorLocator())
ax.get_yaxis().set_minor_locator(mpl.ticker.AutoMinorLocator())
ax.grid(b=True, which='major', color='black', linewidth=0.3, alpha=0.5)
ax.grid(b=True, which='minor', color='black', linewidth=0.3, alpha=0.5)
ax.plot(X,f(X),color="blue")
ax.plot(X1[1:],g(X1[1:]), 'bo',color="red", markersize=8)
#ax.plot(X1,f(X1), '*',color="green")
ax.grid(b=True, which='major', color='black', linewidth=0.3, alpha=0.5)
ax.grid(b=True, which='minor', color='black', linewidth=0.3, alpha=0.5)
ax.plot(X2,l(f(X2)),color="orange")
ax.plot(X3,l(f(X3)), 'bo',color="red", markersize=8)
ax.plot(X2,f(l(X2)),color="orange")
ax.plot(X3,f(l(X3)), 'bo',color="red", markersize=8)
#ax.plot(X3,l(X3), '*',color="green")
plt.plot(X2,l(X2),color="black")
plt.plot(X3,h(X3), 'bo',color="red", markersize=8)
plt.plot(X3,l(X3), 'bo',color="green", markersize=8)
#####################################
ax = plt.subplot(2,2,3)
ax.get_xaxis().set_minor_locator(mpl.ticker.AutoMinorLocator())
ax.get_yaxis().set_minor_locator(mpl.ticker.AutoMinorLocator())
ax.grid(b=True, which='major', color='black', linewidth=0.3, alpha=0.5)
ax.grid(b=True, which='minor', color='black', linewidth=0.3, alpha=0.5)
ax.plot(X2,l(f(X2)),color="orange")
ax.plot(X3,l(f(X3)), 'bo',color="red", markersize=8)
ax.plot(X2,f(l(X2)),color="orange")
ax.plot(X3,f(l(X3)), 'bo',color="red", markersize=8)
#####################################
ax = plt.subplot(2,2,4)
ax.get_xaxis().set_minor_locator(mpl.ticker.AutoMinorLocator())
ax.get_yaxis().set_minor_locator(mpl.ticker.AutoMinorLocator())
ax.grid(b=True, which='major', color='black', linewidth=0.3, alpha=0.5)
ax.grid(b=True, which='minor', color='black', linewidth=0.3, alpha=0.5)
ax.plot(X,k(X),color="blue")
ax.plot(X1[1:],k(X1[1:]), 'bo',color="red", markersize=8)
ax.plot(X3,k(X3), 'bo',color="green", markersize=8)
#ax.plot([4/np.log(2)],k([4/np.log(2)]), 'bo',color="yellow")


plt.show()
```
